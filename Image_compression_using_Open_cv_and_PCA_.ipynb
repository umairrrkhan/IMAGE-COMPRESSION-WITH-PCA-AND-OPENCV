```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
try:
  img = cv2.imread('1.jpg')
  img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
except Exception as e:
  print("Error:", e)
```

```python
img.shape
```

```python
plt.imshow(img)
```

```python
r , g ,b = cv2.split(img)
```

```python
r , g , b =r/255 , g/255 , b/255
```

```python
plt.imshow(r)
```

```python
plt.imshow(b)
```

```python
pca_component = 60

pca_r = PCA(n_components= pca_component)

reduced_r = pca_r.fit_transform(r)

pca_g = PCA(n_components= pca_component)

reduced_g = pca_g.fit_transform(g)

pca_b = PCA(n_components= pca_component)

reduced_b = pca_b.fit_transform(b)
```

```python
combined = np.array([reduced_r,reduced_b,reduced_g])
```

```python
# combined.reshape(4700,3,50).shape
```

```python
reconstructed_r = pca_r.inverse_transform(reduced_r)
reconstructed_g = pca_g.inverse_transform(reduced_g)
reconstructed_b = pca_b.inverse_transform(reduced_b)
```

```python
reconstructed_r.shape
```

```python
img_reconstructed = cv2.merge((reconstructed_r ,reconstructed_b , reconstructed_g))
```

```python

plt.imshow(img_reconstructed)
```
